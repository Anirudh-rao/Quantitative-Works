# 1. Experiments , Outcomes, Sample Spaces

In Probablity, **experiment** is any procedure or process that can be repeated infinitely and has a well-defined set of possible results. Think of flipping a coin, rolling a die measuring a patients tempearture or recording that time  it takes for website to load.
